```python
from sklearn import metrics as metrics
import sklearn
def accuracy_in_samples(clf, X_train, y_train, X_test, y_test, min_train_num=1, num_step=100):
    #TODO number of increment is not implemented
    train_acc = []
    test_acc = []
    train_samples = []
    i = 0
    for num_train in np.linspace(min_train_num, X_train.shape[0], num=num_step):
        i += 1
        num_train = int(num_train)
        print("iteration {}: training size: {}".format(i, num_train), end='\r')

        cX_train = X_train[0:num_train]
        cy_train = y_train[0:num_train]
        
        clf = sklearn.base.clone(clf)
        clf.fit(cX_train, cy_train)
        
        y_ = clf.predict(cX_train)
        train_acc.append(metrics.accuracy_score(cy_train, y_))
        
        y_ = clf.predict(X_test)
        test_acc.append(metrics.accuracy_score(y_test, y_))
        
        train_samples.append(num_train)
    print(len(train_acc))
    print(len(train_samples))
    return train_acc, test_acc, train_samples
        
import matplotlib.pyplot as plt
def plot_accuracy(train_acc, test_acc, train_samples):
    plt.plot(train_samples, test_acc)
    plt.show()
    
train_acc, test_acc, train_samples = accuracy_in_samples(clf, X_train, y_train, X_test, y_test)
plot_accuracy(train_acc, test_acc, train_samples)
    
```

```python
def Ktesting(clf, X_train, y_train, X_val, y_val):
    accuracies = []
    min_val = 1
    max_val = 20
    num_step = 20
    for param_val in np.linspace(min_val, max_val, num=num_step):
        param_val = int(param_val)
        clf = KNeighborsClassifier(n_neighbors=param_val)
        clf.fit(X_train, y_train)
        y_ = clf.predict(X_val)
        accuracies.append(metrics.accuracy_score(y_, y_val))
        print(metrics.accuracy_score(y_, y_val))
    return accuracies
accuracies = Ktesting(clf, X_train, y_train, X_val, y_val)
import pickle

import time
timestr = time.strftime("%Y%m%d-%H%M%S")

pickle.dump(accuracies, open("stats/" + timestr + "-Ktesting.pkl", "wb" ) )
print(accuracies)
```

```python
import itertools
#Taken from sklearn
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    # if normalize:
    #     cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    #     print("Normalized confusion matrix")
    # else:
    #     print('Confusion matrix')


    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```
